## Evolution des salaires suivant le double-diplôme obtenu

Etudions les premières données de l'étude de l'*IESF*, sur l'évolution des salaires en fonction du double-diplôme obtenu. Ces données se résument au tableau suivant :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

salaires = pd.read_csv("dd_salaires.csv", index_col=0)

salaires
```

Précisons de l'*IESF* :

 + *Les écoles d’application des polytechniciens ne sont pas considérées dans ce contexte comme des doubles diplômes*
 + *Les possesseurs d’un Master2 ou DEA ayant poursuivi en thèses sont comptés uniquement comme docteurs.*

### Visualisation des évolutions de salaires

```python
plt.figure(figsize=(8, 6), dpi=100)
plt.plot(salaires, linestyle='--', marker="o", alpha=0.7)
plt.title("Evolution des salaires selon le DD obtenu")
plt.ylabel("Salaires annuels bruts (€)")
plt.xlabel("Catégories d'âge")
plt.legend(salaires.columns, loc='upper left')
plt.savefig("Evolution_salaires_DD.png", transparent=True)
plt.show()
```

```python
histos = salaires.plot.bar(figsize=(10, 8), title="Répartition des salaires par catégorie d'âge, selon le DD obtenu", alpha=0.9)
fig = histos.get_figure()
fig.savefig("Histogramme_salaires_DD.png", transparent=True)
```

### Observations

En début de carrière (<30 ans), on observe un très léger avantage de rémunération pour les DD d'ingénieur en France.

Plus tard, les **DD en management** et les **DD ingénieur à l'étranger** se démarquent nettement. Ainsi, en fin de carrière, le salaire annuel moyen est de *110 000 €* pour ces deux DD, contre *91 800 €* pour le DD d'ingénieur en France et *89 800 €* pour l'ingenieur sans DD.

En ce qui concerne les doctorats, ceux-ci conduisent à des carrières moins bien rémunérées dans le secteur public (seulement *65 375 €* /an en fin de carrière). Dans le secteur privé, ils permettent en fin de carrière une rémunération moyenne supérieure à celle des ingénieurs sans double-diplôme.

### Pistes d'interprétations

Pour le DD à l'étranger, cette tendance peut notamment s'expliquer par le fait que les titulaires d'un second diplôme à l'étranger sont plus amenés à effectuer des carrières internationnales, souvent plus rémunératrices. Quant aux DD en management, ils facilitent l'accès à des postes de direction, également plus rémunérateurs.

## Les formations suivies en seconds diplômes

Etudions à présent le second jeu de données de la brochure de l'*IESF*, concernant les domaines d'activités des ingénieurs, selon le double-diplôme.  Ces données se résument au tableau suivant :

```python
repartitions = pd.read_csv("dd_repartitions.csv", index_col=0)

repartitions
```

### Les différents double-diplômes choisis par les ingénieurs

```python
repartitions['Répartition']

plt.figure(figsize=[10,6], dpi=100)
plt.pie(x=repartitions['Répartition'], autopct="%.1f%%", pctdistance=0.8, explode=[0.1]*len(repartitions))
plt.title("Répartition des DD des ingénieurs", fontsize=14)
plt.legend(repartitions.index, loc='center left', bbox_to_anchor=(1, 0.5), title="Type de second diplôme")
plt.savefig("Repartition_DD.png", transparent=True)
plt.show()
```

Le Master 2 scientifique est le second diplôme de niveau Bac+5 le plus représenté, choisi par 11,8% des ingénieurs, il est suivi par le DD en Management avec 9,8% et le doctorat 7,8%.

Malgré l'important avantage en matière de rémunération du DD Ingénieur à l'étranger (cf. chapitre précédent), celui-ci n'est choisi que par 2,9% de l'ensemble des ingénieurs.

### Les domaines d'activités en fonction des double-diplômes

Regardons à présent si certains double-diplômes semblent faciliter l'accès à certains domaines d'activités.

```python
rep = repartitions.drop("Répartition",axis=1).plot.bar(figsize=(10, 6), title="Répartition des secteurs en fonction du DD",stacked=True).legend(loc='lower center')
rep.get_figure().savefig("Histogramme_domaines_DD.png", transparent=True)
```

On observe deux tendances majeures :

* **48%** des double-diplômes en management mènent une activité commerciale, d'administration, de direction, de conseil, ou d'audit alors que ces activités ne représentent seulement **19%** de celles exercées par l’ensemble des ingénieurs.

* **66%** des doctorats mènent une activité d’étude, R&D ou d’enseignement alors que ces activités représentent seulement **31%** de celles exercées par l’ensemble des ingénieurs.

Pour travailler dans ces deux domaines, les DD en Management (resp. doctorats) semblent donc être les plus pertinents, et à privilégier.

# Conlusion

 * En moyenne, ce sont les DD en management et les DD d'ingénieur à l'étranger qui confèrent les meilleurs rémunération, et cette tendance s'accentue en fin de carrière notamment.
 * Certains DD sont à privilégier si l'on souhaite travailler dans des domaines précis : le doctorat pour les activité d’études, de R&D ou d’enseignement, le DD en Management pour les commerciales, d'administration, de direction, de conseil, ou d'audit.

## Remarques complémentaires

1.  Même si les doctorats conduisent à une carrière moins bien rémunérée dans le secteur public (le plus souvent dans la recherche-enseignement), d'après l'enquête de l'*IESF* :
> *Ces carrières sont en revanche celles qui apportent le plus de satisfaction dans le travail.*
2. Les DD Ingénieur en France peuvent sembler ne pas apporter de différences significatives en matière de rémunération par rapport aux ingénieurs sans DD. Cependant, il ne faut pas oublier qu'ils restent très pertinents dans le cas de spécialisations. Par exemple, selon la formation initiale, un DD avec *Supaéro* peut être la voie royale pour travailler dans l'aéronautique, un DD avec l'*ENSAE* pour travailler dans les data, etc.
3. Cette étude se penche sur les avantages en matière de rémunération qu'offrent les DD, mais il est important de rappeller que ce critère ne devrait pas être le principal élément pour votre choix.
4. Cette étude est réalisé sur des données moyennes, et les interprétations sont donc à prendre avec précaution.
